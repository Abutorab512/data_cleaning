```python
data['Size'] = data['Size'].replace('Varies with device','0')
```

```python
data['Size'] = data['Size'].apply(lambda x: x.replace('+','') if ('+') in str(x) else x)
```

```python
data['Size'].unique()
```

```python
data['Size'] = data['Size'].apply(lambda x: x.replace(',',''))
```

```python
data['Size'] = data['Size'].apply(lambda x: x.replace('k', '') if ('k') in str(x) else x)
```
